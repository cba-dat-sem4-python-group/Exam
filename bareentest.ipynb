```python
from itertools import permutations

def make_royal_flush():
    res = []
    royals = permutations([1, 13, 12, 11, 10])
    for x in royals:
        for i in range(1,5):
            hand = [i,x[0],i,x[1],i,x[2],i,x[3],i,x[4]]
#             print(hand)
            res.append(hand)
    return res

def make_straight_flush():
    res = []
    combi = [[i,i+1,i+2,i+3,i+4] for i in range(1,10)]
    for x in combi:
        perms = permutations(x)
        for y in perms:
            for i in range(1,5):
                hand = [i,y[0],i,y[1],i,y[2],i,y[3],i,y[4]]
#                 print(hand)
                res.append(hand)
    return res

def make_four_of_a_kind():
    res = []
    suits = [1,2,3,4]
    perms = permutations(suits)
    for x in perms:
        for y in suits:
            for a in range(1,14):
                for b in range(1,14):
                    if b != a:
                        res.append([x[0],a,x[1],a,x[2],a,x[3],a,y,b])
    return res

def make_full_house():
    res = []
    
    return res

def make_flush():
    res = []
    
    return res

def make_straight():
    res = []
    
    return res

def make_three_of_a_kind():
    res = []
    
    return res

def make_two_pairs():
    res = []
    
    return res

def make_one_pair():
    res = []
    
    return res

def make_nothing():
    res = []
    
    return res
```
