```python
import random

def get_random_royal_flush():
    suit = random.choice(range(1,5))
    ranks = [1,10,11,12,13]
    random.shuffle(ranks)
    return ranks+[suit,suit,suit,suit,suit,9]

def get_random_straight_flush():
    suit = random.choice(range(1,5))
    low_card = random.choice(range(1,10))
    ranks = [low_card,low_card+1,low_card+2,low_card+3,low_card+4]
    random.shuffle(ranks)
    return ranks+[suit,suit,suit,suit,suit,8]

def get_random_foak():
    rank = random.choice(range(1,14))
    suits = [1,2,3,4]
    random.shuffle(suits)
    last_suit = random.choice(range(1,5))
    last_rank = rank
    while rank == last_rank:
        last_rank = random.choice(range(1,14))
    cards = [(rank,suits[0]),(rank,suits[1]),(rank,suits[2]),(rank,suits[3]),(last_rank,last_suit)]
    random.shuffle(cards)
    return [cards[0][0],cards[1][0],cards[2][0],cards[3][0],cards[4][0],cards[0][1],cards[1][1],cards[2][1],cards[3][1],cards[4][1],7]

def get_random_full_house():
    rank1 = random.choice(range(1,14))
    rank2 = rank1
    while rank1 == rank2:
        rank2 = random.choice(range(1,14))
    suits1 = []
    for _ in range(3):
        while True:
            randomSuit = random.choice(range(1,5))
            if randomSuit not in suits1:
                suits1.append(randomSuit)
                break
    suits2 = []
    for _ in range(2):
        while True:
            randomSuit = random.choice(range(1,5))
            if randomSuit not in suits2:
                suits2.append(randomSuit)
                break
    random.shuffle(suits1)
    random.shuffle(suits2)
    cards = [(rank1,suits1[0]),(rank1,suits1[1]),(rank1,suits1[2]),(rank2,suits2[0]),(rank2,suits2[1])]
    random.shuffle(cards)
    return [cards[0][0],cards[1][0],cards[2][0],cards[3][0],cards[4][0],cards[0][1],cards[1][1],cards[2][1],cards[3][1],cards[4][1],6]

def get_random_flush():
    suit = random.choice(range(1,5))
    ranks = []
    for _ in range(5):
        while True:
            randomCard = random.choice(range(1,14))
            if randomCard not in ranks:
                ranks.append(randomCard)
                break
    random.shuffle(ranks)
    return ranks+[suit]*5+[5]

def get_random_straight():
    low_card = random.choice(range(1,11))
    ranks = [low_card,low_card+1,low_card+2,low_card+3,1 if low_card==10 else low_card+4]
    suits = []
    for _ in range(5):
        while True:
            randomCard = random.choice(range(1,5))
            if suits.count(randomCard) < 4:
                suits.append(randomCard)
                break
    random.shuffle(ranks)
    random.shuffle(suits)
    return ranks+suits+[4]

def get_random_toak():
    rank = random.choice(range(1,14))
    suits = [1,2,3,4]
    random.shuffle(suits)
    ex_suits = []
    for _ in range(2):
        ex_suits.append(random.choice(range(1,5)))
    ranks = []
    for _ in range(2):
        while True:
            randomCard = random.choice(range(1,14))
            if randomCard not in ranks and randomCard != rank:
                ranks.append(randomCard)
                break
    random.shuffle(ranks)
    random.shuffle(ex_suits)
    cards = [(rank,suits[0]),(rank,suits[1]),(rank,suits[2]),(ranks[0],ex_suits[0]),(ranks[1],ex_suits[1])]
    random.shuffle(cards)
    return [cards[0][0],cards[1][0],cards[2][0],cards[3][0],cards[4][0],cards[0][1],cards[1][1],cards[2][1],cards[3][1],cards[4][1],3]

def get_random_two_pairs():
    rank1 = random.choice(range(1,14))
    rank2 = rank1
    while rank1 == rank2:
        rank2 = random.choice(range(1,14))
    rank3 = rank1
    while rank1 == rank3 or rank2 == rank3:
        rank3 = random.choice(range(1,14))
    suits1 = []
    for _ in range(2):
        while True:
            randomSuit = random.choice(range(1,5))
            if randomSuit not in suits1:
                suits1.append(randomSuit)
                break
    suits2 = []
    for _ in range(2):
        while True:
            randomSuit = random.choice(range(1,5))
            if randomSuit not in suits2:
                suits2.append(randomSuit)
                break
    suit3 = random.choice(range(1,5))
    random.shuffle(suits1)
    random.shuffle(suits2)
    cards = [(rank1,suits1[0]),(rank1,suits1[1]),(rank2,suits2[0]),(rank2,suits2[1]),(rank3,suit3)]
    random.shuffle(cards)
    return [cards[0][0],cards[1][0],cards[2][0],cards[3][0],cards[4][0],cards[0][1],cards[1][1],cards[2][1],cards[3][1],cards[4][1],2]

def get_random_pair():
    rank = random.choice(range(1,13))
    suits = []
    for _ in range(2):
        while True:
            randomSuit = random.choice(range(1,5))
            if randomSuit not in suits:
                suits.append(randomSuit)
                break
    ranks = []
    for _ in range(3):
        while True:
            randomCard = random.choice(range(1,14))
            if randomCard not in ranks and randomCard != rank:
                ranks.append(randomCard)
                break
    ex_suits = [random.choice(range(1,5)) for _ in range(3)]
    random.shuffle(suits)
    random.shuffle(ranks)
    random.shuffle(ex_suits)
    cards = [(rank,suits[0]),(rank,suits[1]),(ranks[0],ex_suits[0]),(ranks[1],ex_suits[1]),(ranks[2],ex_suits[2])]
    random.shuffle(cards)
    return [cards[0][0],cards[1][0],cards[2][0],cards[3][0],cards[4][0],cards[0][1],cards[1][1],cards[2][1],cards[3][1],cards[4][1],1]

def get_random_nothing():
    ranks = []
    for _ in range(5):
        while True:
            randomCard = random.choice(range(1,14))
            if randomCard not in ranks:
                ranks.append(randomCard)
                break
    suits = []
    for _ in range(5):
        while True:
            randomSuit = random.choice(range(1,5))
            if suits.count(randomSuit) < 4:
                suits.append(randomSuit)
                break
    random.shuffle(ranks)
    random.shuffle(suits)
    res = ranks+suits+[0]
    return res if hp.check_straight(res) else get_random_nothing()
```
